## Matplotlib

```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

# Load an example image from scikit-image
img_array = np.array(data.astronaut())  # Example astronaut image
plt.imshow(img_array)
plt.axis("off")
plt.title("astronaut")
plt.show()
```

```python
img_array
print(img_array.shape)
```

**`skimage`**: Provides sample images; data.astronaut() loads an example astronaut image.

**`numpy:`** Handles array operations, which we use to manipulate the image as a matrix.

**`matplotlib.pyplot`**: Used to display images for visualization.


### Cropping the Image

```python
# Crop a 100x100 region from the top-left corner
cropped_img = img_array[:100, :100]  # Selects rows and columns up to index 100
plt.imshow(cropped_img)
plt.title("Cropped Image (100x100)")
plt.axis('off')
plt.show()
```

**Crop Image:** `img_array[:100, :100]` selects the top-left 100 rows and 100 columns, creating a 100x100 pixel crop.


**Display Cropped Image:** The cropped image is displayed, and axes are hidden for clarity.

### Remove the Red Channel

```python
img_no_red = img_array.copy()
img_no_red[:, :, 0] = 0
plt.imshow(img_no_red)
plt.title("Image without Red Channel")
plt.axis('off')
plt.show()
```

Create a Copy: img_array.copy() creates a copy of the image to avoid modifying the original.

Remove Red Channel: img_no_red[:, :, 0] = 0 sets the red channel to 0 for all pixels.

[:, :, 0] selects the first color channel (red) across all rows and columns.

Assigning 0 to this channel removes the red tones, resulting in an image where only green and blue colors contribute to each pixel.

Display Modified Image: The altered image without the red channel is shown.

```python
# Flip the Image Horizontally
flipped_img = np.flip(img_array, axis=0)  # Flips the image horizontally
plt.imshow(flipped_img)
plt.title("Horizontally Flipped Image")
plt.axis('off')
plt.show()
```

Horizontal Flip: np.flip(img_array, axis=1) flips the image along the horizontal axis.

axis=0 specifies the direction of the flip (left to right).

Every pixel in each row is reversed, creating a mirror image.

```python
# Rotate the Image by 90 Degrees
rotated_img = np.rot90(img_array)  # Rotates the image 90 degrees counterclockwise
plt.imshow(rotated_img)
plt.title("Rotated Image (90 Degrees)")
plt.axis('off')
plt.show()
```
